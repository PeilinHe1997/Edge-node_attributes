```python
import networkx as nx
import numpy as np
import pandas as pd
```

```python
G1 = nx.Graph()
G1.add_edges_from([(0,1),(0,2),(0,3),(0,5),(1,3),(1,6),(3,4),(4,5),
                  (4,7),(5,8),(8,9)])

nx.draw_networkx(G1)
```

## Adjacency List

```python

G_adjlist = open(r"G_adjlist.txt","a")
for line in nx.generate_adjlist(G1):
    print(line)
    G_adjlist.write(line+"\n")
G_adjlist.close()

```

```python
G2 = nx.read_adjlist("G_adjlist.txt",nodetype = int)
G2.edges()
```

## Adjacency Matrix

```python
G_mat = np.array([[0,0,1],[1,0,0],[1,1,0]])
G3 = nx.Graph(G_mat)
G3.edges()
```

## Edgelist

```python
G_edgelist = open(r"G_edgelist.txt","a")
edgelist = ["0 1 4",
"0 2 5",
"0 3 2",
"0 5 6",
"1 3 2",
"1 6 5",
"3 4 3",
"5 4 1",
"5 8 2",
"4 7 6",
"8 9 1"]
for i in edgelist:
    G_edgelist.write(i+"\n")
G_edgelist.close()
```

```python
G4 = nx.read_edgelist("G_edgelist.txt",data = [("weigt",int)])
```

```python
G4.edges(data = True)
```

## Pandas DataFrame

```python
G_df = pd.read_csv("G_edgelist.txt",delim_whitespace = True, header = None, names = ["n1","n2","weight"])

G_df
```

```python
G5 = nx.from_pandas_edgelist(G_df,"n1","n2",edge_attr = "weight")
nx.draw_networkx(G5)
```
