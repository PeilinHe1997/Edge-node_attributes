```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
pop = [30,345,232,564,234,57,234,67,234,56]
for i in range(0,10):
    G.add_node(i,population = pop[i])

G.add_edges_from([(0,1),(0,2),(0,3),(0,5),(1,3),(1,6),(3,4),(4,5),
                  (4,7),(5,8),(8,9)])
attr = {}
weight = [3,4,5,5,3,4,5,5,3,4,4]
for i in range(0,len(G.edges())):
    attr[list(G.edges())[i]] = weight[i]
nx.set_edge_attributes(G, attr,"weight")
```

```python
G.edges(data = True)
```

```python
plt.figure(figsize = (10,9))
nx.draw_networkx(G)
```

```python
plt.figure(figsize = (10,9))
pos = nx.circular_layout(G)
nx.draw_networkx(G,pos)
```

```python
plt.figure(figsize = (10,7))
# set attributes

node_color = [G.degree(v) for v in G]
node_size = [ 10 * nx.get_node_attributes(G,"population")[v] for v in G]
edge_width = [2 * G[u][v]["weight"] for u,v in G.edges()]

nx.draw_networkx(G,pos, alpha = 0.7, with_labels = False, edge_color = ".4", node_size = node_size, width = edge_width, cmap = plt.cm.Blues)
plt.axis("off")
plt.tight_layout()
```

```python
plt.figure(figsize = (10,7))
nx.draw_networkx(G,pos, alpha = 0.7, with_labels = False, edge_color = ".4", node_size = node_size, width = edge_width, cmap = plt.cm.Blues)
## highlight some important nodes
greater_than_300  = [x[0] for x in G.nodes(data = True) if x[1]["population"] > 300]
nx.draw_networkx_nodes(G, pos, nodelist = greater_than_300, node_color = "red")

plt.axis("off")
plt.tight_layout()
```
