```python
import networkx as nx
```

```python
G = nx.Graph()
G.add_edges_from([("A","K"),("A","B"),("A","C"),("B","C"),("B","K"),("C","E"),("C","F"),('D',"E"),("E","F"),("E","H"),("F","G")])
```

```python
nx.draw_networkx(G)
```

A undirected graph is connected if, for every pair of nodes, there is a path between them.

```python
nx.is_connected(G)
```

```python
G1 = nx.Graph()
G1.add_edges_from([("A","K"),("A","B"),("A","C"),("B","C"),("B","K"),('D',"E"),("E","F"),("E","H"),("F","G")])
```

```python
nx.draw_networkx(G1)
```

### Connected component in undirected graph:
A subset of nodes such as:

1) Every node in the subset has a path to every other node.

2) No other node has a path to any node in the subset.

```python
# number of connected components in graph
nx.number_connected_components(G1)
```

```python
# nodes of subset(connected components)
sorted(nx.connected_components(G1))
```

```python
# nodes of a specific connected component that includes node "F"
nx.node_connected_component(G1,"F")
```

### Connected component in directed graph:

##### Strongly connected
A directed graph is strongly connected if, for every pair nodes u and v, there is a directed path from u to v and a directed path from v to u.


```python
G2 = nx.DiGraph()
G2.add_edges_from([("A","K"),("A","B"),("A","C"),("B","C"),("B","K"),("C","E"),("C","F"),('D',"E"),("E","F"),("E","H"),("F","G")])
nx.draw_networkx(G2)
```

```python
nx.is_strongly_connected(G2)
```

##### Wealy connected:
A directed graph is weakly connected if replacing all directed edges with undirected edges produces a connected undirected graph.

```python
nx.is_weakly_connected(G2)
```

##### Strongly connected component:
A subset of nodes such as:

1) Every node in the subset has a directed path to every other nodes.

2) No other node has a directed path to and from every node in the subset.

```python
G3 = nx.DiGraph()
G3.add_edges_from([("A","B"),("A","C"),("B","C"),("B","K"),("K","A"),('D',"E"),("C","K"),("E","F"),("E","H"),("F","G")])
nx.draw_networkx(G3)
```

```python
nx.is_weakly_connected(G3)
```

```python
sorted(nx.strongly_connected_components(G3))
```

#### Weakly connected component:
The connected components of the graph after replacing all directed edges with undirected edges.

```python
sorted(nx.weakly_connected_components(G3))
```
